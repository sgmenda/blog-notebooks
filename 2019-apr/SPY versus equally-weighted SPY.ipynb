```python
from quantopian.research import prices, symbols
import pandas as pd
```

```python
spy_close = prices(
    assets=symbols('SPY'),
    start='2004-1-1',
    end='2019-4-1',
)
```

```python
rsp_close = prices(
    assets=symbols('RSP'),
    start='2004-1-1',
    end='2019-4-1',
)
```

Say I bought \$1000 of SPY and RSP on Jan 2, 2013.

```python
spy_amount = 1000/spy_close[0]
rsp_amount = 1000/rsp_close[0]
```

```python
pd.DataFrame({
    '$1000 SPY': spy_amount*spy_close,
    '$1000 RSP': rsp_amount*rsp_close,
}).plot(
    title="$1000 in SPY versus equally-weighted SPY"
);
```

```python
print("On 1 Apr, 2019, I would have")
print("SPY: $%d" % (spy_amount*spy_close[-1]))
print("RSP: $%d" % (rsp_amount*rsp_close[-1]))
```
